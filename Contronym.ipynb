```python
import gensim
```

```python
pretrained_embeddings_path = "https://s3.amazonaws.com/dl4j-distribution/GoogleNews-vectors-negative300.bin.gz"
word2vec = gensim.models.KeyedVectors.load_word2vec_format(pretrained_embeddings_path, binary=True)
```

```python
word = ['strike', 'seed', 'trim','bound','fast','clip','sanction','screen','weather','bolt','consult','custom','dust','finished','out','off','overlook','left', 'mature']
meaning_1 = ['hit', 'add', 'add','moving','secured','fasten','allow','hide','withstand','separate','give','common','sprinkle','completed','visible','nonfunctional','monitor','remaining', 'appropriate']
meaning_2 = ['miss', 'remove', 'remove','restrained','quick','remove','ban','show','wear','hold','get','unique','clean','destroyed','invisible','functional','neglect','departed','inappropriate']
```

```python
import matplotlib.pyplot as plt
import scipy
from scipy.spatial import distance
```

```python
for index in range(len(word)):
  word_meaning_1_similarity = 1 - scipy.spatial.distance.cosine(word2vec[word[index]], word2vec[meaning_1[index]])
  word_meaning_2_similarity = 1 - scipy.spatial.distance.cosine(word2vec[word[index]], word2vec[meaning_2[index]])
  similarity_sum = word_meaning_1_similarity + word_meaning_2_similarity
  meaning_labels = [meaning_1[index], meaning_2[index]]
  plt.pie([word_meaning_1_similarity/similarity_sum, word_meaning_2_similarity/similarity_sum], labels = meaning_labels, autopct='%1.1f%%')
  plt.title(word[index])
  plt.show()
```
